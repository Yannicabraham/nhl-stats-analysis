### Explanation: Columns
- RK = Rank/ID
- Player = Name
- Age = Age during Season
- Pos = Position
- Tm = Team
- GP = Games Played
- G = Goals
- A = Assists
- PTS = Points

- PIM = Penalty Infraction Minutes, number of min. a player has been assessed 
- PS = Point Shares, number to estimate number of standing points a player contributes to the team 
- EV = Even Strength Goals
- PP = Power Play Goals
- SH = Short Handed Goals
- GW = Game Winning Goals / a way to recognize the player who scored the goal that effectively secured the victory for their team. 
- EV.1 = Even Strength Assists
- PP.1 = Power Play Assists
- SH.1 = Short Handed Assists
- S = Shots on Goals
- S_percent = Shooting Percentage / percentage of shots on goal that result in a goal being scored. S% = Goals / Shots on Goal
- TOI = Time on Ice 
- ATOI = Average Time on Ice = TOI/GP
- BLK = Blocked Shots
- HIT = Hits 
- FOW = Faceoffs Won
- FOL = Faceoffs Lost
- FO_percent = measures a player's effectiveness in winning faceoffs. FO% =  number of faceoffs won / total number of faceoffs taken
- HART = Binary 1/0. 1 = Winner of the Hart Memorial Trophy in a year. MVP Award for regular season.
- Votes = Count of Votes for the Hart Memorial Trophy
- Season = Season

```python
#Players by position
plt.figure(figsize=(12,6))
sns.countplot(x=df['Pos'], data=df, palette='pastel')
plt.title('Player Count per Position')
plt.xlabel('Position Abbreviation')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

print("The dataset contains over 4000 D-Zone players, which is the position with the most players. Given the fact that the defensive line of a team consists of a couple of D-men, this only makes sense. Players with Left Wing and Right Wing as position balance each other out. The Center\nposition proves to be a vital one too with over 3500 players in the dataset.\nThe hybrid positions only make up a small fraction of all positions.")
```

```python
#Blocks by position
blocks_pos = df.groupby('Pos')['BLK'].sum().sort_values(ascending=False)
blocks_pos = blocks_pos[blocks_pos > 1000]
blocks_pos.to_frame()
```

```python
blocks_pos.plot(kind='bar', color='firebrick')
plt.title('Shots Blocked by Position')
plt.xlabel('Position')
plt.ylabel('Shots Blocked')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("As expected, D-Men block the most shots in the NHL, followed by Center Forwards with significantly less blocks. In the next step,\nwe're analyzing the Shots Blocked by Position over the seasons. Have teams and players adapted their behavior, has it become harder to block shots?")
```

```python
#Overall goals by season
goals_per_season = df['G'].groupby(df['Season']).sum()
goals_per_season.plot(kind='bar', color='firebrick')
plt.title('Goals Count per Season')
plt.xlabel('Season')
plt.ylabel('Goal Count')
plt.grid(True, alpha=0.3)
plt.show()

print("The 2013 season was shortened due to a lockout, which explains the drop in goals. Afar from that, a pretty constant count in goals during every season is observed.")
```

```python
#Power Play goals over the seasons
pp_goals_trend = df.groupby('Season')['PP'].sum().sort_values().to_frame()
pp_goals_trend.sort_values('Season')
```

```python
pp_goals_trend.sort_values('Season').plot(kind='bar', color='firebrick')
plt.title('Power Play Goals per Season')
plt.xlabel('Season')
plt.ylabel('Power Play Goals')
plt.grid(True, alpha=0.3)
plt.show()

print("Interestingly, play off goals have decreased significantly over seasons. Since the 2006 campaign saw over 2500 PP goals, the number of goals in even-strength situations have decreased to now roughly 1500 per season in 2018. This is likely due to a combination of factors, including increased defensive structure, imporoved PK strategies, and better goaltending. Again, the dip in the 2013 campaign is exlpained by the shortended season.")
```

```python
#Top10 Overall Goal Scorers
#(1)Nach Saison und Player gruppieren und die Tore aufsummieren
goals_by_season = df.groupby(['Season','Player'])['G'].sum().reset_index()

#(2)F√ºr jede Saison Top Goal Scorer ermitteln
top_scorer_by_season = goals_by_season.sort_values(by=['Season','G'], ascending=[True, False])
top_scorer_by_season = top_scorer_by_season.groupby('Season').head(1)
top_scorer_by_season = top_scorer_by_season[['Season','Player','G']]
top_scorer_by_season
```

```python
#Visualizing
plt.figure(figsize=(8,6))
sns.barplot(data=top_scorer_by_season, x='Season', y='G', hue='Player')
plt.title('Top Goal Scorer by Season')
plt.ylabel('Amount Goals')
plt.xlabel('Season')
plt.legend(loc='upper right')
plt.tight_layout()
plt.grid(True, alpha=0.3)
plt.show()

print("The top goal scorer visual delivers insightful findings. 4-time MVP Alex Ovechkin has scored the most goals per season in 7 different years! An outstanding performance. Also, Ovechkin's 2008 campaign with his 65 goal record, did not reflect in his MVP votes we analyzed above. That year Ovechkin received only 888 votes, despite his strongest goal-scoring season.")
```

```python
#Most Power Play Goals by Season (Player Level)
#(1)Nach Saison und Player gruppieren und PPG summieren
ppg_by_season = df.groupby(['Season','Player'])['PP'].sum().reset_index()
ppg_by_season

#(2)F√ºr jede Saison Top PPG Scorer ermitteln
top_ppg_by_season = ppg_by_season.sort_values(by=['Season','PP'], ascending=[True, False])
top_ppg_by_season = top_ppg_by_season.groupby('Season').head(1)
top_ppg_by_season = top_ppg_by_season[['Season','Player','PP']]
top_ppg_by_season
```

```python
#Visualizing
plt.figure(figsize=(8,6))
sns.barplot(data=top_ppg_by_season, x='Season', y='PP', hue='Player')
plt.title('Most Power Play Goals per Year')
plt.ylabel('Power Play Goals')
plt.tight_layout()
plt.grid(True, alpha=0.3)
plt.legend(loc='upper right')
plt.show()

print("The most PPG by a player vary significantly across seasons. Stressing Ovechkin's status in the NHL, he also had the most PPG in 6 different \nseason.")
```

```python
#Plusminus distribution
print("In ice hockey, the plus‚Äìminus measures a player's goal differential. When a team that is at even-strength or shorthanded scores a goal, all\nplayers on the ice on the scoring team will register a plus while all players on the conceding team on the ice will register a minus.\n")

df.plusminus.hist(bins=10, edgecolor='white', color='coral')
plt.xlabel('Plusminus Stat')
plt.title('Distribution of Plusminus Stats\n', loc='left')
plt.grid(True, alpha=0.3)
plt.show()

print("The distribution reveals that most players (approx. 6000) have only a slight positive to slight negative plusminus statistic. A very strong\nplusminus can only be observed for a small fraction of players (combined approx. 1000).")
```

```python
#Distribution of TOI depending on Plusminus stat
#Clustering plusminus in groups
def plusminus_group(row):
    if row.plusminus >= 50:
        return '+50s'
    elif row.plusminus >= 40:
        return '+40s'
    elif row.plusminus >= 30:
        return '+30s'
    elif row.plusminus >= 20:
        return '+20s'
    elif row.plusminus >= 10:
        return '+10s'
    elif row.plusminus >= 0:
        return '+00s'
    elif row.plusminus > -10:
        return '-00s'
    elif row.plusminus > -20:
        return '-10s'
    elif row.plusminus > -30:
        return '-20s'
    elif row.plusminus > -40:
        return '-30s'
    elif row.plusminus > -50:
        return '-40s'

df['plusminus_group'] = df.apply(plusminus_group, axis=1)
df.head()
```

```python
toi_by_plusminus = df.groupby(by='plusminus_group').TOI.mean()
toi_by_plusminus.to_frame()
```

```python
#Visualizing with barchart
toi_by_plusminus.plot(kind='bar', color='firebrick')
plt.xlabel('Plus Minus Cluster')
plt.ylabel('Average TOI')
plt.grid(True, alpha=0.3)
plt.xticks(rotation=45)
plt.show()

print("Tentatively, the more extrem the plusminus stat (either positive or negative), the higher the average time on ice. Considering that a player with plenty of TOI has more opportunities to affect his plusminus (both positive and negative), this only makes sense.")
```
