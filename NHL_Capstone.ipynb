### Explanation: Columns
- RK = Rank/ID
- Player = Name
- Age = Age during Season
- Pos = Position
- Tm = Team
- GP = Games Played
- G = Goals
- A = Assists
- PTS = Points
- plusminus = '+' indicates player was on ice when they scored even-strength or short-handed, '-' indicates player was on ice when they conceded a goal.
- PIM = Penalty Infraction Minutes, number of min. a player has been assessed 
- PS = Point Shares, number to estimate number of standing points a player contributes to the team 
- EV = Even Strength Goals
- PP = Power Play Goals
- SH = Short Handed Goals
- GW = Game Winning Goals / a way to recognize the player who scored the goal that effectively secured the victory for their team. 
- EV.1 = Even Strength Assists
- PP.1 = Power Play Assists
- SH.1 = Short Handed Assists
- S = Shots on Goals
- S_percent = Shooting Percentage / percentage of shots on goal that result in a goal being scored. S% = Goals / Shots on Goal
- TOI = Time on Ice 
- ATOI = Average Time on Ice = TOI/GP
- BLK = Blocked Shots
- HIT = Hits 
- FOW = Faceoffs Won
- FOL = Faceoffs Lost
- FO_percent = measures a player's effectiveness in winning faceoffs. FO% =  number of faceoffs won / total number of faceoffs taken
- HART = Binary 1/0. 1 = Winner of the Hart Memorial Trophy in a year. MVP Award for regular season.
- Votes = Count of Votes for the Hart Memorial Trophy
- Season = Season

```python
#Overall goals per season
goals_per_season = df['G'].groupby(df['Season']).sum()
goals_per_season.plot(kind='bar', color='firebrick')
plt.title('Goals Count per Season')
plt.xlabel('Season')
plt.ylabel('Goal Count')
plt.grid(True, alpha=0.3)
plt.show()

print("The 2013 season was shortened due to a lockout, which explains the drop in goals. Afar from that, a pretty constant count in goals during every season is observed.")
```

```python
#Players by position
plt.figure(figsize=(12,6))
sns.countplot(x=df['Pos'], data=df, palette='pastel')
plt.title('Player Count per Position')
plt.xlabel('Position Abbreviation')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

print("The dataset contains over 4000 D-Zone players, which is the position with the most players. Given the fact that the defensive line of a team consists of a couple of D-men, this only makes sense. Players with Left Wing and Right Wing as position balance each other out. The Center\nposition proves to be a vital one too with over 3500 players in the dataset.")
```

```python
#Penalty Minutes by position
print("The condition PIM > 100 serves to filter out small fractions that affect the below pie chart.")
penalty_by_pos = df['PIM'].groupby(df['Pos']).sum().sort_values(ascending=False).to_frame()
penalty_by_pos = penalty_by_pos[penalty_by_pos > 100]
penalty_by_pos
```

```python
penalty_by_pos.plot(kind='pie', subplots=True, startangle=180, autopct='%1.1f%%', labels=df['Pos'].value_counts().index)
plt.legend(title='Position', loc='upper right')
plt.tight_layout()
plt.title('Penalty Minutes Share by Position')
plt.ylabel('')
plt.show()

print("35% of all Penalty Minutes were accumulated by defense players. Offensive players almost level each other out, with 20% for wingers and\nalmost 24% for centers.")
```

```python
#Plusminus distribution
print("In ice hockey, the plus‚Äìminus measures a player's goal differential. When a team that is at even-strength or shorthanded scores a goal, all\nplayers on the ice on the scoring team will register a plus while all players on the conceding team on the ice will register a minus.\n")

df.plusminus.hist(bins=10, edgecolor='white', color='coral')
plt.xlabel('Plusminus Stat')
plt.title('Distribution of Plusminus Stats\n', loc='left')
plt.grid(True, alpha=0.3)
plt.show()

print("The distribution reveals that most players (approx. 6000) have only a slight positive to slight negative plusminus statistic. A very strong\nplusminus can only be observed for a small fraction of players (combined approx. 1000).")
```

```python
#Distribution of TOI depending on Plusminus stat
#Clustering plusminus in groups
def plusminus_group(row):
    if row.plusminus >= 50:
        return '+50s'
    elif row.plusminus >= 40:
        return '+40s'
    elif row.plusminus >= 30:
        return '+30s'
    elif row.plusminus >= 20:
        return '+20s'
    elif row.plusminus >= 10:
        return '+10s'
    elif row.plusminus >= 0:
        return '+00s'
    elif row.plusminus > -10:
        return '-00s'
    elif row.plusminus > -20:
        return '-10s'
    elif row.plusminus > -30:
        return '-20s'
    elif row.plusminus > -40:
        return '-30s'
    elif row.plusminus > -50:
        return '-40s'

df['plusminus_group'] = df.apply(plusminus_group, axis=1)
df.head()
```

```python
toi_by_plusminus = df.groupby(by='plusminus_group').TOI.mean()
toi_by_plusminus.to_frame()
```

```python
#Visualizing with barchart
toi_by_plusminus.plot(kind='bar', color='firebrick')
plt.xlabel('Plus Minus Cluster')
plt.ylabel('Average TOI')
plt.grid(True, alpha=0.3)
plt.xticks(rotation=45)
plt.show()

print("Tentatively, the more extrem the plusminus stat (either positive or negative), the higher the average time on ice. Considering that a player with plenty of TOI has more opportunities to affect his plusminus (both positive and negative), this only makes sense.")
```
